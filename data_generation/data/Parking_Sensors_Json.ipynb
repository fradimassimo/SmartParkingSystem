```python
#Libraries
import json
import random
from datetime import datetime
```

```python
# Function to generate a random date in ISO 8601 format (with random year, month, and day)
def generate_random_date():
    # Randomly select year (2023 or 2024)
    year = random.choice([2023, 2024])

    # Randomly select month between 1 and 12
    month = random.randint(1, 12)

    # Get the number of days in the selected month
    days_in_month = {
        1: 31, 2: 29 if year % 4 == 0 else 28, 3: 31, 4: 30, 5: 31, 6: 30,
        7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31
    }

    # Randomly select a valid day in the chosen month
    day = random.randint(1, days_in_month[month])

    # Generate the random date (use the chosen year, month, and day)
    random_date = datetime(year, month, day)

    # Convert the date to ISO 8601 format (with arbitrary time and timezone offset)
    iso_format_date = random_date.isoformat() + "T00:00:00+01:00"  # Fixed time and timezone offset

    return iso_format_date

def generate_random_item():
    return {
        "lastupdated": generate_random_date(),
        "status_timestamp": generate_random_date(),
        "zone_number": round(random.uniform(1,10000)), 
        "status_description": random.choice(["Present","Unoccupied"]), 
        "kerbsideid": round(random.uniform(1,10000)), 
        "location": {"lon": round(random.uniform(100,200),14), "lat": round(random.uniform(-50,0),14)}
        }
```

```python
# Function to generate multiple random items
def generate_multiple_items(n):
    return [generate_random_item() for _ in range(n)]

# Example: Generate 10 random items
new_data = generate_multiple_items(10)

# Print or save the extended JSON data
with open('extended_data.json', 'w') as f:
    json.dump(new_data, f, indent=4)

print(json.dumps(new_data, indent=4))  # Print the extended JSON to the console
```
